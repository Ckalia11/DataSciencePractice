```python
import pandas as pd
```

```python
# strings need to be within [] brackets
pd.DataFrame({'name': ['chetan'], "age" : 19}, index = ['f'])
```

```python
pd.DataFrame({'name': ['chetan', 'raj'], "age" : [19, 20]}, index = ['a', 'b'])
```

```python
# pass a list
pd.Series(['a','b','c', 'd'])
```

```python
data = pd.read_csv('wine.csv')
```

```python
# iloc takes indicies
data.iloc[:, 1:4].head(3)
```

```python
# loc can take strings
# returns all columns from country to designation
data.loc[0:2, 'country' : 'designation']
```

```python
# gets only the two columns
data.loc[:, ['country', 'designation']]
```

```python
# can pass indices as a list
data.loc[[1,2,5], ['country', 'price']]
```

```python
data.head(4)
```

```python
# set a column of the df as an index for easier access
# in order to access points for US
data.loc[3, 'price']
```

```python
# instead set country as index and then use loc to avoid index
data.set_index('country').loc['US', 'points']
```

```python
# get all rows that have a country of Italy and a price of 15.0
data.loc[(data.country == 'Italy') & (data.price == 15)].head(3)
```

```python
data.loc[(data.country == 'Italy') & (data.price.isnull())].head(3)
```

```python
# get all records where country is Italy or Portugal
data.loc[data.country.isin(['Italy', 'Portugal'])].head(3)
```
