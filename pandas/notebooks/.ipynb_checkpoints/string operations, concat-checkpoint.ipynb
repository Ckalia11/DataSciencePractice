```python
import pandas as pd
```

```python
df = pd.read_csv('USA_cars_datasets.csv')
df.set_index(['brand', 'model'])
df.sort_index()
df.head()
```

```python
lst = df.condition.str.split(" ")
lst




```

```python
df.loc[df.condition.isin(['10 days left'])].head()
```

```python
hours = []
hours_index = []
days = []
days_index = []
for i in range(0, len(lst)):
    if lst[i][1] == 'hours':
        hours.append(lst[i][0])
        hours_index.append(i)
    else:
        days.append(lst[i][0])
        days_index.append(i)

```

```python
hours1 = pd.Series(hours, index = hours_index, name = 'hours')
days1 = pd.Series(days, index = days_index, name = 'days')
```

```python
days1
```

```python
df = pd.concat([df, hours1, days1], axis = 1)
df.head()
```

```python
df['country'] = df.country.str.strip()
df.set_index(['country', 'year'], inplace = True)
df.sort_index()
```

```python
df['condition1'] = df.condition.str.split(" ").apply(lambda a : "".join(a))
df.condition1
```
