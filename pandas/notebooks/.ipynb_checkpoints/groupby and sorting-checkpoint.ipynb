```python
import pandas as pd
df = pd.read_csv('../data/wine.csv')
df.head(3)
```

```python
# for each designation, get the minimum price
df.groupby('designation').price.min()
```

```python
df.groupby(['country', 'province']).agg({'points': 'mean'})
```

```python
# for each country, get the mean, max, and min price using the agg function
df.groupby('country').price.agg(['mean', 'max', 'min']).head()
```

```python
# new df storing min price for each country and designation
df1 = df.groupby(['country', 'designation']).price.min()
df1.head(10)
df1.tail()
```
