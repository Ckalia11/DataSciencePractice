```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
df = pd.DataFrame(np.random.rand(50,3), columns = ['a', 'b', 'c'])
df.head()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=df)

plt.figure(figsize=(15,5))
sns.lineplot(data = df['b'], label = 'b')
sns.lineplot(data = df['c'], label = 'c')
```

```python
plt.figure(figsize=(15,5))
sns.set_style('whitegrid')
plt.title('The graph of c against index')
sns.barplot(x = df.index, y = df['c'])
plt.ylabel('column c')
```

```python
# annot TRUE gives numerical values
sns.heatmap(data = df.iloc[:15, :], annot = True)
```

```python
sns.scatterplot(x = df['a'], y = df['b'])
```

```python
sns.regplot(x = df['a'], y = df['b'])
```

```python
lst = []
for i in range(50):
    lst.append('yes')
    lst.append('no')
d = pd.Series(lst)
df['answer'] = d
df.head()
```

```python
sns.scatterplot(x = df['a'], y = df['b'], hue = df['answer'])
```

```python
sns.lmplot(x = 'a', y = 'b', hue = 'answer', data = df)
```

```python
sns.swarmplot(x = 'answer', y = 'a', data = df)
```

```python
# frequency of values
plt.figure(figsize = (10,5))
plt.title('Frequency distribution of column values')
sns.distplot(a = df['a'], label = 'frequency of a values', kde = False)
sns.distplot(a = df['b'], label = 'frequency of b values', kde = False)
#force legend
plt.legend()
```

```python
plt.figure(figsize= (10,5))
plt.title('The density plot!')
sns.kdeplot(data = df['a'], shade = True)
sns.kdeplot(data = df['b'], shade = True)
sns.kdeplot(data = df['c'], shade = True)
```
