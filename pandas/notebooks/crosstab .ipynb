```python
import pandas as pd
```

```python
df = pd.read_csv('../data/USA_cars_datasets.csv')
df.head()
```

```python
# use crosstab to see frequency between two features
pd.crosstab(df.brand, df.country, margins = True).head()
```

```python
# can also do for multiple 
pd.crosstab([df.brand, df.model], df.country, margins = True)
```
