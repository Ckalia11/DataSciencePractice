```python
import pandas as pd
```

```python
df = pd.read_csv('../data/yelp.csv')
df.head()
```

```python
def func(s):
    lower = []
    upper = []
    for i in s:
        if i.islower():
            lower.append(i)
        elif i.isupper():
            upper.append(i)
    return (lower, upper)
```

```python
m = df.review_id.apply(func)
df['lower'] = m.apply(lambda a: a[0])
df.head()
df.isnull().sum()
df.head()
df.loc[:, 'date':'text']
```

```python
class person:
    def __init__(self, name, age, skin_color):
        self.name = name
        self.age = age
        self.skin_color = skin_color
    def print_items(self, gender):
        if gender == 'm':
            print('{} is {} and his skin color is {}'.format(self.name, self.age, self.skin_color))
        else :
            print('{} is {} and her skin color is {}'.format(self.name, self.age, self.skin_color))
    
        
p1 = person('chetan', 19, 'brown')
p1.print_items('m')
p1.name = 'julie'
p1.print_items('f')

```
