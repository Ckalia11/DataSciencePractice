```python
df1 = pd.DataFrame({'employee': ['Bob', 'Chetan', 'Lisa', 'Jerry'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})
```

```python
df1
```

```python
df2
```

```python
# inner only selects employee names common between both df's
pd.merge(df1, df2, how = 'inner')
```

```python
# outer get's them all
pd.merge(df1, df2, how = 'outer')
```

```python
# gets names from left df
pd.merge(df1, df2, how = 'left')
```

```python
# gets names from right df
pd.merge(df1, df2, how = 'right')
```

```python
# if column names aren't the same
df1.rename(columns={"employee": "names"}, inplace = True)
df1.head()
```

```python
pd.merge(df1, df2, left_on = 'names', right_on = 'employee', how = 'outer').head()
```
