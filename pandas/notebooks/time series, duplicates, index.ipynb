```python
df = pd.read_csv('http://bit.ly/uforeports')
df.head()
```

```python
# time series 
df.isnull().sum()
df.dropna(subset = ['Colors Reported']).isnull().sum()
df['hour'] = df.Time.str.split(' ').apply(lambda a : a[1])
df['date'] = df.Time.str.split(' ').apply(lambda a : a[0])
df.loc[:, ['Time', 'date']]
df['day'] = df.date.str.split('/').apply(lambda a: a[0])
df['month'] = df.date.str.split('/').apply(lambda a: a[1])
df['year'] = df.date.str.split('/').apply(lambda a: a[2])
df.loc[:, ['date', 'day', 'month', 'year', 'hour']]
df.drop('date', axis = 1, inplace = True)
df.head()
```

```python
# date time
df['Time'] = pd.to_datetime(df.Time)
df.head()
```

```python
df.Time.dt.day
```

```python
df.Time.dt.hour
```

```python
# can compare two datetime objects
t = pd.to_datetime(10/5/1930)
df.loc[df.Time >= t]
```
